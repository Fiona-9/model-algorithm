# KMeans 聚类
<line height=2.8em>

    一种典型的无监督学习算法，主要用于将相似的样本自动归到一个类别中。
    聚类算法中根据样本之间的相似性，将样本划分到不同的类别中，对于不同的相似度计算方法，会得到不同的聚类结果，常用的相似度计算方法有欧式距离法。

## 简单演示

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans,MiniBatchKMeans
from sklearn.metrics import calinski_harabasz_score,silhouette_score
```

```python
# 随机创建数据集
x,y = make_blobs(n_samples=1000,n_features=2,centers=[[2,2],[-1,-3],[4,5]],cluster_std=[0.5,1,0.8])
# 散点图 可视化
plt.figure(figsize=(20,8))
plt.scatter(x[:,0],x[:,1])
```

```python
# 模型训练
estimator = KMeans(n_clusters=3,random_state=3)
estimator.fit(x)
y_pre = estimator.fit_predict(x)
# 数据可视化
plt.figure(figsize=(6,4))
plt.scatter(x[:,0],x[:,1],c=y_pre)
plt.show()
# 模型评估---CH系数的值越大越好
print("CH系数：\n",calinski_harabasz_score(x,y_pre))
```

## 算法流程

<line height=3.8em>
    

+ K : 初始中心点个数（计划聚类数）
    

+ means：求中心点到其他数据点距离的平均值  

    
+ 时间复杂度 $O(n*k*I)$

### <font color = brown>**聚类步骤**</font>：
1. 随机设置K个聚类空间的点作为初始的聚类中心


2. 对于其他的每个点，计算到K个中心的距离，位置的点选择最近的一个聚类中心点作为标记类别


3. 重新计算出每个聚类的新中心点（平均值）


4. 如果质心不再移动，结束程序，否则重新进入第二步过程


5. 当每次迭代结果不变时，认为算法收敛，聚类完成，**K-Means一定会停下，不可能陷入一直选质心的过程**。

### <font color = brown>**优缺点评价**</font>：
  + 由于每次都要计算所有的样本与每一个质心之间的相似度，故**在大规模的数据集上，K-Means算法的收敛速度比较慢**。
  
   
   + K-Means算法对噪声和离群点敏感（中心点易偏离）
   
   
   + 结果不一定是全局最优，只能保证**局部最优**。对初始值敏感，不同的初始值会导致不同的结果。
  
  
  + 很难发现大小和密度差别很大的簇并进行增量计算

## 原理和模型评估
<font size = 4>  
$ cost~ function~(SSE~ /~The~ sum~ of ~squares ~due ~to ~error)) $
    $$ SSE = \sum_{i=1}^k \sum_p \mid p-m_i \mid^2~~(p\in C_i) $$
</font>

  SSE的大小，衡量了聚类的松散程度

### <font color = brown>**“肘”方法：**</font>
1. 对于n个点的数据集，迭代计算 k from 1 to n，每次聚类完成后计算SSE


2. SSE是会逐渐变小的，直到k==n时平方和为0，因为每个点都是它所在的簇中心本身。


3. 在这个平方和变化过程中，会出现一个拐点也即“肘”点，**下降率突然变缓时即认为是最佳的k值**。在决定什么时候停止

   训练时，肘形判据同样有效，数据通常有更多的噪音，在增加分类无法带来更多回报时，我们停止增加类别。

```python
# 创建数据集
x,y = make_blobs(n_samples=1000,n_features=2,centers=[[2,2],[-1,-3],[4,5]],cluster_std=[0.5,1,0.8])
# 设置循环的模型训练
SSE = []
for i in range(1,10):
    transfer = KMeans(n_clusters=i,random_state=15)
    transfer.fit(x)
    SSE.append(transfer.inertia_)
print(SSE)
# 数据可视化
plt.figure(figsize=(10,6))
plt.scatter(range(1,10),SSE,color="orange")
plt.plot(range(1,10),SSE)
plt.grid(linestyle="--", alpha=0.5)
plt.show()
```

由上图可见，k=3 时情况比较理想

### <font color = brown>**轮廓系数法（Silhouette Coefficient）**</font>

    结合了聚类的凝聚度（Cohesion）和分离度（Separation），用于评估聚类的效果
  
<font size = 4> 
$$ S_i = \frac{b_i-a_i}{max(a_i,b_i)}  $$
</font>

a_i:簇内不相似度,样本i到同簇其他样本的平均距离  

b_i:样本i到其他某簇Cj 的所有样本的平均距离bij，称为样本i与簇Cj 的不相似度。定义为样本i的簇间不相似度：

<font size = 4> 
    $$bi =min\left\{bi1, bi2, ..., bik\right\}$$
</font>
    
所有样本的$S_i$的均值称为聚类结果的轮廓系数$S$，是该聚类是否合理、有效的度量。  

**$S$接近于1时**，聚类实现了**内部距离最小化，外部距离最大化**

### <font color = brown>**CH系数（Calinski-Harabasz Index）**</font>

类别内部数据的距离平方和越小越好，类别之间的距离平方和越大越好-----分数S高则聚类效果好

<font size = 4> 
$$S(k)= \frac{SSB}{SSW} \frac{m-k}{k-1}$$
$$SSW(每簇内各点距中心点的平方和)= \sum_{i=1}^m(x_i-C_i)^2$$  
$$SSB（每簇内中心点距所有数据平均值点的平方和）= \sum_{j=1}^k(C_j-\overline{X})^2$$  
</font>

其中，$\frac{m-k}{k-1}$ 表示用尽量少的类别聚集尽量多的样本，$m$为训练集样本数，$k$为类别数




```python
# 创建数据集
x,y = make_blobs(n_samples=1000,n_features=2,centers=[[2,2],[-1,-3],[4,5]],cluster_std=[0.5,1,0.8])
# 设置循环的模型训练
CH = []
for i in range(2,10):
    transfer = KMeans(n_clusters=i,random_state=15)
    transfer.fit(x)
    y_pre = transfer.fit_predict(x)
    CH.append(calinski_harabasz_score(x,labels=y_pre))
# 数据可视化
plt.figure(figsize=(10,6))
plt.scatter(range(2,10),CH,color="orange")
plt.plot(range(2,10),CH)
plt.grid(linestyle="--", alpha=0.5)
plt.show()
```

由上图可见，k=3 时,CH系数最大，情况比较理想

## 优化算法

###  <font color = brown>**Canopy+kmeans（Canopy粗聚类配合kmeans）**</font>

+ 帮助确定最优的K值


+ 仍有t1、t2的确定问题


**算法流程：**
1. 随机选取一点（作为中心点），分别以t1、t2为半径做圆，t1圆内点属于一类


2. 另随机选取t2圆外一点，重复上述步骤，直至所有点都有归属

```python
import math
import random
import numpy as np
from datetime import datetime
from pprint import pprint as p
import matplotlib.pyplot as plt

# 随机生成一个500个二维 [0,1)平面点
dataset = np.random.rand(500, 2)
# print(dataset)

# 设计一个Canopy类
class Canopy:
    def __init__(self, dataset):
        self.dataset = dataset
        self.t1 = 0
        self.t2 = 0
    
    # 设置初始阈值t1 和 t2
    def setThreshold(self, t1, t2):
        if t1 > t2:
            self.t1 = t1
            self.t2 = t2
        else:
            print("t1 needs to be larger than t2!")
    
    # 使用欧式距离进行距离计算
    def euclideanDistance(self, vec1, vec2):
        return math.sqrt(((vec1 - vec2) ** 2).sum())
    
    # 根据当前dataset的长度随机选择一个下标
    def getRandIndex(self):
        return np.random.randint(len(self.dataset))
        # return random.randint(0, len(self.dataset) - 1)
    
    # 核心算法
    def clustering(self):
        if self.t1 == 0:
            print('Please set the threshold t1 and t2!')
        else:
            canopies = [] # 用于存放最终归类的结果
            while len(self.dataset) != 0:
                # 获取一个随机下标
                rand_index = self.getRandIndex() 
                # 随机获取一个中心点，定为P点
                current_center = self.dataset[rand_index] 
                # 初始化P点的canopy类容器
                current_center_list = [] 
                # 初始化P点的删除容器
                delete_list = [] 
                # 删除随机选择的中心点P
                self.dataset = np.delete(self.dataset, rand_index, 0)
                for datum_j in range(len(self.dataset)):
                    datum = self.dataset[datum_j]
                    # 计算选取的中心点P到每个点之间的距离
                    distance = self.euclideanDistance(current_center, datum)
                    if distance < self.t1:
                        # 若距离小于t1，则将点归入P点的canopy类
                        current_center_list.append(datum)
                    if distance < self.t2:
                        # 若小于t2则归入删除容器
                        delete_list.append(datum_j)
                self.dataset = np.delete(self.dataset, delete_list, 0)
                canopies.append((current_center, current_center_list))
            return canopies
        
        
def showCanopy(canopies, dataset, t1, t2):
    fig = plt.figure(figsize=(10, 10))
    
    sc = fig.add_subplot(111)
    
    colors = ['brown', 'green', 'blue', 'y', 'r', 'tan', 'dodgerblue', 'deeppink', 'orangered', 'peru', 'blue', 'y', 'r',
              'gold', 'dimgray', 'darkorange', 'peru', 'blue', 'y', 'r', 'cyan', 'tan', 'orchid', 'peru', 'blue', 'y', 'r', 'sienna']
    markers = ['*', 'h', 'H', '+', 'o', '1', '2', '3', ',', 'v', 'H', '+', '1', '2', '^',
               '<', '>', '.', '4', 'H', '+', '1', '2', 's', 'p', 'x', 'D', 'd', '|', '_']
    for i in range(len(canopies)):
        canopy = canopies[i]
        center = canopy[0]
        components = canopy[1]
        sc.plot(center[0], center[1], marker=markers[i], color=colors[i], markersize=10)
        t1_circle = plt.Circle(
            xy=(center[0], center[1]), radius=t1, color='dodgerblue', fill=False)
        t2_circle = plt.Circle(
            xy=(center[0], center[1]), radius=t2, color='skyblue', alpha=0.2)
        sc.add_artist(t1_circle)
        sc.add_artist(t2_circle)
        for component in components:
            sc.plot(component[0], component[1], marker=markers[i], color=colors[i], markersize=1.5)
    
    maxvalue = np.amax(dataset)
    minvalue = np.amin(dataset)

    # print('maxvalue = ', maxvalue)
    # print('minvalue = ', minvalue)
    # print('t1 = ', t1)
    # print('t2 = ', t2)
    # print(minvalue - t1, maxvalue + t1)
    plt.axis('equal')
    plt.xlim((minvalue - t1, maxvalue + t1))
    plt.ylim((minvalue - t1, maxvalue + t1))
    # plt.axis('scaled')
    
    
    plt.savefig('Canopy.svg')
    plt.show()
    
def main():
    t1 = 0.6
    t2 = 0.4
    gc =Canopy(dataset)
    gc.setThreshold(t1, t2)
    canopies = gc.clustering()
    print('Get %s initial centers.' % len(canopies))
    showCanopy(canopies, dataset, t1, t2)
    
if __name__ == "__main__":
    main()
```

###  <font color = brown>**kmeans++（距离越远越容易成为新的质心）**</font>

<font size=4 >
$$ P = \frac{D(x)^2}{\sum_{x \in X}D(x)^2}$$
</font>

+ 让KMeans的结果尽量趋近于全局最优，选择的初始中心点尽量地远


**算法流程：**
1. 从输入的数据点集合（要求有k个聚类）中随机选择一个点作为第一个聚类中心


2. 对于数据集中的每一个点x，计算它与最近聚类中心(指已选择的聚类中心)的最短距离D(x)


3. 选择一个新的数据点作为新的聚类中心，选择的原则是：D(x)较大的点，被选取作为聚类中心的概率较大，P为点x被选择的概率


4. 重复2和3直到k个聚类中心被选出来


5. 利用这k个初始的聚类中心来运行标准的k-means算法


###  <font color = brown>**二分k-means（拆除SSE最大的簇）**</font>

+ 每一步都会抑制掉最大的误差-----早发现早治疗


+ 不怎么受初始化程度的影响


**算法流程：**
1. 开始所有的点作为一个簇


2. 选择能最大程度降低SSE的簇（SSE较大的簇）划分为两个簇


3. 一直进行下去，知道簇的数目等于用户输入的K值为止

###  <font color = brown>**k-medoids（和kmeans选取中心点的方式不同）**</font>

+ 相比之下，对异常值、噪声没有那么敏感


+ 使用中位数作为聚点，聚点为实际的数据点


+ 对初始值的依赖性不强


+ 对小规模的数据效果更好，时间复杂度$O(n^2*k*I)$


+ 需要提前设定K值


+ 有可能陷入局部最优解，但是比Kmeans的结果更为稳定


**算法流程**


1. 确定聚类的个数K。


2. 在所有数据集合中选择K个点作为各个聚簇的中心点。


3. 计算其余所有点到K个中心点的距离（欧式距离），并把每个点到K个中心点最短的聚簇作为自己所属的聚簇。


4. 在每个聚簇中按照顺序依次选取点，计算该点到当前聚簇中所有点距离（曼哈顿距离）之和，最终距离之后最小的点，则视为新的中心点。


5. 重复（2），（3）步骤，直到各个聚簇的中心点不再改变。

###  <font color = brown>**kernel kmeans（映射到高维空间）**</font>

+ 处理线性不可分的问题（如环状聚类），从低维到高维的映射（在高维空间可以做到线性可分），在进行KMeans聚类


+ 会用到一些高维映射函数，例如高斯和函数


+ 复杂度高，计算量大

### <font color = brown>**ISODATA（动态聚类）**</font>

+ 类别数目随着聚类过程为变化

  1. 类别数目的合并：某一类中样本数太少，或两个类间的距离太近
  2. 类别数目的分裂：类内的方差太大
  
  
 + 最终输出的聚类范围：$[\frac{k}{2},2k]$
 
 
 + 常用于卫星图的聚类

### <font color = brown>**Mini-batch K-Means（大数据集分批聚类）**</font>

+ 每次迭代时，**从不同类别样本中抽取一部分样本**来更新质心

+ 处理大数据时，时间和效果更优


`estimator = MiniBatchKMeans(n_clusters=8,batch_size=100,random_state=10)`其中，batch_size为每次迭代选取样本点的个数
